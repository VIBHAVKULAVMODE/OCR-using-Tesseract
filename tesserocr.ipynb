```python
from tesserocr import PyTessBaseAPI
from PIL import Image
import cv2
import os
import pytesseract
import matplotlib.pyplot as plt
```

```python
images=['p1.jpeg','p2.jpg','p3.jpg','p4.jpg','p5.jpg','p6.jpeg']
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'
```

```python
for img in images:
    
    text = pytesseract.image_to_string(Image.open(img))
    
    imgs = cv2.imread(img)
    plt.imshow(imgs, cmap = 'gray', interpolation = 'bicubic')
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.show()
    print(text)
```

```python
with PyTessBaseAPI(path='C:/users/HP/Downloads/tesserocr-master/tessdata-master/tessdata-master', lang='eng') as api:
    for img in images:
        imgs = cv2.imread(img)
        plt.imshow(imgs, cmap = 'gray', interpolation = 'bicubic')
        plt.xticks([]), plt.yticks([]) 
        plt.show()
        api.SetImageFile(img)
        print(api.GetUTF8Text())
        print(api.AllWordConfidences())
```
